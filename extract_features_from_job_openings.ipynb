```python
def fillAttribute(word, temp):
    vagas = [x.lower() for x in df_vagas.TITLE]
    requisitos = [x.lower() for x in df_vagas.REQUIREMENTS]
    atividades = [x.lower() if type(x) == str else '' for x in df_vagas.ACTIVITIES]
    
    for ds in [vagas, requisitos, atividades]:
        for index, row in enumerate(ds):
            if word in row:
                temp[index] = 1
    return temp


df = {}
for key in keywords:
    temp = np.zeros(len(df_vagas), dtype=int)
    for word in keywords[key]:
        fillAttribute(word, temp)
    df[key] = temp
    
df = pd.DataFrame(df)
```

```python
df1 = pd.concat((df, pd.get_dummies(df_vagas.LEVEL_ID)), axis=1)
df1.rename(columns={1:'estagiario', 2: 'junior', 3: 'pleno', 4: 'senior'}, inplace=True)
df1 = pd.concat((df1, pd.get_dummies(df_vagas.DEPARTMENT_ID)), axis=1)
df1.rename(columns={1:'dsa', 2: 'dsb', 3: 'dsc', 4: 'dsd', 5: 'dse'}, inplace=True)
df1 = pd.concat((df1, pd.get_dummies(df_vagas.SITE_ID)), axis=1)
df1.rename(columns={1:'bsb', 2: 'cps', 3: 'man', 4: 'poa'}, inplace=True)
df1 = pd.concat((df1, pd.get_dummies(df_vagas.TITLE)), axis=1)
df1.tail()
```

```python
from kmed import kmedoids
from scipy.spatial.distance import squareform, pdist
from scipy.spatial.distance import correlation

dist_matrix = squareform(pdist(df, 'correlation'))

medoids, values = kmedoids(dist_matrix, 5)
map_cluster = {cluster: values[idx] for idx, cluster in enumerate(medoids)}
```

```python
df_vagas.iloc[39]
```

```python
def recommender(requirements_vector):
    
    dist_list = []
    for med in medoids:
        dist_list.append((correlation(requirements_vector, df1.iloc[med]), med))
    
    cluster = sorted(dist_list)[0][1]
    
    print(cluster)
    
    recommendation = []
    for job in map_cluster[cluster]:
        recommendation.append((correlation(requirements_vector, df1.iloc[job]), df_vagas.iloc[job].TITLE, job))
    
    return [job[1] for job in sorted(recommendation)][:10]


requirements_vector = df1.iloc[39]
recommender(requirements_vector)
```

```python
dict_req = {key: 0 for key in df1.columns}

dict_req['java'] = 1
dict_req['android'] = 1
dict_req['pleno'] = 1
dict_req['cps'] = 1
dict_req['dsa'] = 1
# dict_req['computer_vision'] = 1

recommender(list(dict_req.values()))
```
